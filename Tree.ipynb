# Decision Tree Classifier

A decision Tree is a ml model which follows a set of if-else conditions to visualize the data and classify it according to the conditions.

Think of array in which data are organized linearly but in tree it happens in non-linear way.

9. **Branch or Sub Tree -** It is a tree made from part of the entire tree by splitting from a branch.

10. **Pruning -** Removing a certain part of tree from a subnode

### Entropy

In simple terms, It is a measure of disorder,impurity, uncertainity or randomness of data.

In decision Tree, Entropy controls on process of splitting of data. It affects how a decision tree draw its boundary.

It is measured between 0 & 1(but depending on your number of classes it can be greater than 1).

---
**Lower the Entropy Higher the Certainity and Vice Versa**

---

The Equation of Entropy E is 

$H(S) = -\sum_{i=1}^{c} P_ilog_2(P_i)$

**WHERE,**

$H(S) $ is the Entropy obtained

$c $ is the number of classes to be classified

$P_i$ is the Frequentist Probability of Classification of an class 'i' in our data

**FOR EG:** In above example, we had 10 sample datas in which 5 of them belonged to 'YES' class and 5 of them to 'NO' class.

$P(YES) = \frac{No. \ of \ data \ points \ belonging \ to \ YES \ class}{Total \ Data \ Points} = \frac{5}{10} = 0.5$


$P(NO) = \frac{No. \ of \ data \ points \ belonging \ to \ NO \ class}{Total \ Data \ Points} = \frac{5}{10} = 0.5$

$H(S) = -\sum_{i=1}^{2} P_ilog_2(P_i)$ 

$\ \ \ \ \ \ \ \ \  = -P_{YES}log_2(P_{YES}) -P_{NO}log_2(P_{NO}) $

$\ \ \ \ \ \ \ \ \ = -0.5log_2(0.5) -0.5log_2(0.5)$ 

$\ \ \ \ \ \ \ \ \ = 1 $

**FOR EG:** In another example, we had 20 sample datas in which 7 of them belongs to '+' class and 13 of them belongs to '-' class.

$P(+) = \frac{No. \ of \ data \ points \ belonging \ to \ + \ class}{Total \ Data \ Points} = \frac{7}{20} = 0.35$


$P(-) = \frac{No. \ of \ data \ points \ belonging \ to \ - \ class}{Total \ Data \ Points} = \frac{13}{20} = 0.65$

$H(S) = -\sum_{i=1}^{2} P_ilog_2(P_i)$ 

$\ \ \ \ \ \ \ \ \  = -P_-log_2(P_-) -P_+log_2(P_+) $

$\ \ \ \ \ \ \ \ \ = -0.65log_2(0.65) -0.35log_2(0.35)$ 

$\ \ \ \ \ \ \ \ \ = 0.93 $

### Information Gain



Information Gain is defined as difference between the original information requirement(i.e based on proportion of classes) and the new requirement (i.ee obtained after partitioning on a feature column)


Information Gain tells us how much would be gained by branching on a feature column.

---
It is given by formula

$Gain(A) = Entropy(S) - Entropy_A(S)$

**WHERE**,
$Entropy_A(S) = \sum_{c \in A} P(c)E(c)$ is the entropy after partitioning on feature column A



Let us consider a below caseimgonline-com-ua-twotoone-MeD34KRfhDzRIkvp.jpg

Our split at X = 2 produces two branches

1. Left Branch with 4 brown

2. Right Branch with 5 Yellow and 1 Brown

WE can see that our split isn't optimal, but we can determine how good is our split with the help of Information Gain.


Before the split, we had 5 yellow and 5 brown , so our Entropy was

$Entropy_{before} = -(0.5log_20.5+0.5log_20.5) = 1$

---

After the split, we have two branches:

In left Branch,

$P_{left} = \frac{4}{10} = 0.4$

$E_{left} = -(\frac{4}{4}log_2(\frac{4}{4}) + \frac{0}{4}log_2(\frac{0}{4})) =0$, because it's a dataset of all one color.


In Right Branch,

$P_{left} = \frac{6}{10} = 0.6$

$E_{right} =-(\frac{1}{6}log_2(\frac{1}{6}) + \frac{5}{6}log_2(\frac{5}{6})) = 0.65 $

---

Now, Let us Find the Entropy after Partitioning.

$E_{split} = \sum_{c \in 2} P(c)E(c) = 0.4*0 + 0.6*0.65 = 0.39 $

As, you see the parent node had Entropy $E_{before} = 1$ while after the split the Children Enropy $E_{split} = 0.39$. 

**Information Gain = how much Entropy we removed**

Information Gain = $E_{before} - E_{split} = 1 -0.39 = 0.61$

In perfect sense, each branch would contain only one color after the split,which would be zero entropy **i.e** $E_{split} = 0$

### Gini Index
It measures the degree of probability of a particular variable being wrongly classified when it is randomly choosen.

It is also known as Gini Impurity.

It is given by

$Gini = 1 -\sum_{i=1}^{c}(P_i)^2$
**WHERE** c is the number of classes and $P_i$ is the probability of an object being classified to a particular class

Gini Index starts from 0 to 1 where 0 = Complete Equality , 1 = Complete Inequality and 0.5 = Equally Distributed

**Now it's time to get into dive into Decision Tree Algorithms**

### 1. Iterative Dichotomiser 3 (ID3) 
**Uses Entropy Function and Information Gain as metrics**

ID3 uses a top-down greddy approach to build a decision tree. The top down refers to starting to build the tree from the tree and the greedy approach means that at each iteration we select the best feature at the present moment to create a node (i.e No Backtracking)

#### Algorithm for ID3 -

1. We need to calculate the entropy of the target class.

2. The dataset is then split on the different attributes. The entropy for each branch is calculated. Then it is added proportionally, to get total entropy for the split. The resulting entropy is subtracted from the entropy before the split. The result is the Information Gain, or decrease in entropy.

3. Choose attribute with the largest information gain as the decision node, divide the dataset by its branches and repeat the same process on every branch.

4. 

  a. A branch with entropy of 0 is a leaf node.

  b. A branch with entropy more than 0 needs further splitting.

5. The ID3 algorithm is run recursively on the non-leaf branches, ubtil all data is classified.

**CREDIT** [ID3 Algorithm](https://www.saedsayad.com/decision_tree.htm)

### 2. CART (Classification and Regression Tree)

1. Calculate weighted sum of gini indexes for all nominal features.

2. Compare the Gini indices score, after n before using new attribute to separate data. If the node itself has the lowest score, than there is no point in separating the data

3. If separating the data result in an improvement, than pick the separation with the lowest impurity score

# IRIS FLOWER CLASSIFICATION

```
#Importing all the basic necessities
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Importing Datasets
from sklearn.datasets import load_iris

#Importing Helper Functions 
from sklearn.model_selection import train_test_split

#Importing our today's model
from sklearn.tree import DecisionTreeClassifier

#Importing Performance Metrics
from sklearn.metrics import confusion_matrix, accuracy_score , f1_score 
```

```
#Loading our datasets
dataset = load_iris()
print(dataset.keys())
```

```
#Saving our features matrix and target vector respectively
X = dataset['data']
y = dataset['target'] 
column_names = dataset['feature_names']
target_class = dataset['target_names']
```

```
#Knowing about target classes and features names
print(column_names)
print('\n')
print(target_class)
```

```
#Splitting the dataset
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size = 0.8)
```

```
#Fitting the model
classifier = DecisionTreeClassifier(criterion='entropy')
classifier.fit(x_train,y_train)

```

```
# Plotting our tree
from sklearn import tree
tree.plot_tree(classifier)
```

```
#Predicting the model
y_pred = classifier.predict(x_test)


#Looking out into metrics
cn=confusion_matrix(y_test,y_pred)
acc = accuracy_score(y_test,y_pred)
 

print('Confusion Matrix is ',cn)
print('Accuracy is ',acc)
```

# Random Forest

**Multiple trees make a forest and choosing a random tree from that forest is Random Forest**

It is an ensemble method or known as Ensemble Modelling.

### Ensemble Method
Ensemble is combining many different independent machine learning models to arrive at a decision. The decision is taken by majority vote of these models. Each of the constituent models can be just better than a random guess, but by probability, the chances of over half of these models giving wrong results at a time is extremely low. This is the reason why ensemble methods consisting of very simple decision trees give comparable results to most complex models that exist.

An example of Ensemble Learning is Random Forest. It also helps to reduce overfitting on the Decision Trees.

ensemble.png

Random forest basically is the model made up of combination of various decision trees built on random samples having different policies for splitting.

Random Forest uses following topics to make up its algorithm: 

1. [Decision Tree Method](https://medium.com/towards-artificial-intelligence/decision-trees-explained-with-a-practical-example-fe47872d3b53)

3. [Ensemble Learning](https://www.analyticsvidhya.com/blog/2015/08/introduction-ensemble-learning/)

4. [Cross Validation](https://machinelearningmastery.com/k-fold-cross-validation/)  
  


```
#Random Forest Implementation
from sklearn.ensemble import RandomForestClassifier

randomForestClassifier = RandomForestClassifier(n_estimators=100, criterion ='gini', max_depth=5)
randomForestClassifier.fit(x_train,y_train)
y_pred = randomForestClassifier.predict(x_test)


#Looking out into metrics
cn=confusion_matrix(y_test,y_pred)
acc = accuracy_score(y_test,y_pred)
 

print('Confusion Matrix is ',cn)
print('Accuracy is ',acc)
```
