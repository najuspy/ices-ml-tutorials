```python
x = np.arange(0,3*np.pi,0.01)
y= np.sin(x)
z= np.cos(x)

plt.plot(x,y,'ys-')
plt.plot(x,z,'go--')

plt.title('Various Waves Representation',fontsize=20)
plt.xlabel('Frequency')
plt.ylabel('Value')

plt.xlim([0,8])
plt.ylim([-2,2])

plt.xticks(np.arange(0, 3.5*np.pi, np.pi/2  ),['0','/2','','3/2','2','5/2','3','3/2','4']) 


plt.legend(['Sine','Cosine'])

plt.grid()
plt.show()
```

### *Important --* 

The thing above we don't use it now. We use concept of Object Oriented Plotting and create **figure** and **axes** because of its versatility and easiness while creating plots.

**Figure** :-  It is the topmost layer of the plot (kind of big-picture). Figure constitutes of subplots, sub axis, titles, subtitles, legends, everything inside the plot but an overview.

**Axes** :-   Its a part of the Figure, nothing but a subplot
Axes define a subplot, we can write our own x-axis limits, y-axis limits, their labels, the type of graph. It controls every detail inside the subplot.

**Basics Step in Creating Plot in Matplotlib**

1. Prepare Data
2. Create Plot
3. Plot
4. Customize Plot
5. Save Plot (Optional)
6. Show Plot

```python
# Creation of same above plot using object oriented method

# step 1  (Prepare Data)
x = np.arange(0,3*np.pi,0.01)
y= np.sin(x)
z= np.cos(x)

# step 2 (Create Plot)
fig = plt.figure()   #This is top level container of all plot elements. We're initializing a figure object using figure method.
ax = fig.add_axes([0,0,1,1]) #Axes create a reactangle inside figure as parameterized by [left,bottom,width,height]. 

# step 3 (Plot)
l1 = ax.plot(x,y,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x,z,'go--') # dash line with green colour and circle marker

# step 4 (Customize Plot)
ax.set_title('Various Waves Representation',fontsize=20)
ax.set_xlabel('Frequency')
ax.set_ylabel('Value')

ax.set_ylim([-2,2])

ax.legend(labels = ('Sine', 'Cosine'), loc = 'lower right') # legend placed at lower right

ax.grid(color='r', linestyle='-', linewidth=0.5) #Grid of red color 

#Step 5 (Saving Plot)
'''
plt.savefig('waves.png',transparent=True) 
'''

#Step 6 (Show Plot)
plt.show()





```

## Subplots 

Sometimes we don't want to stack different plot in same figure and want to have a grouping of figure. We will be using subplot for that particular purpose.

The `subplot()` function is use to create number of subplots inside a figure.

SYNTAX --  `fig, ax = plt.subplots(nrows,ncols,index)`


At first lets see why we require subplots.

```python
# Plot of Square vs Cube
x = np.arange(100)
y= x ** 2
z= x ** 3

fig = plt.figure()
ax= fig.add_axes([0,0,1,1])

l1 = ax.plot(x,y , label='Square')
l2= ax.plot(x,z, label='Cube')

ax.set_title('Cube vs Square')
ax.set_xlabel('Values')
ax.set_ylabel('Processed Values')
ax.legend()

plt.show()
```

We can see that both graph aren\'t in same scale and it is difficult to visualize in that context.

Now lets us subplot in this case.

```python
x = np.arange(0,15)
y= x ** 2
z= x ** 3

print('Horizontally Stacked')
fig,ax = plt.subplots(1,2) #1 rows and 2 columns
ax[0].plot(x,y)
ax[1].plot(x,z)
plt.show() 


```

```python
print('Vertically Stacked')
fig,ax = plt.subplots(2,1) #2 rows and 1 columns
ax[0].plot(x,y)
ax[1].plot(x,z)
plt.show()
```

Its getting fun right. Let us add some more into it. Make a (2,2) figure.

```python
print('Matrix Stacked') 
fig,ax = plt.subplots(3,2) #3 rows and 2 columns

ax[0,0].plot(x,y)   
ax[0,0].set_title('X vs Y')

ax[0,1].plot(y,x) 
ax[0,1].set_title('Y vs X')

ax[1,0].plot(x,z)   
ax[1,0].set_title('X vs Z')

ax[1,1].plot(z,x)   
ax[1,1].set_title('Z vs X')

ax[2,0].plot(y,z)   
ax[2,0].set_title('Y vs Z')

ax[2,1].plot(z,y)   
ax[2,1].set_title('Z vs Y')


plt.show()
```

# 2D PLOT

IN 2D plots, we either plot various independent variables with each other or we plot each independent variable with respect to dependent variable. 

There are many types of 2D plots each with various use cases.

##  Correlation Plots
The Correlation plots are used to visualize the relationship between 2 or more variables. That is, how does one variable change with respect to another.

### Scatter Plot
A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.
scatter .png

```python
x = np.random.randint(0,100,200)
y = np.random.randint(0,100,200)
fig,ax =plt.subplots()
ax.scatter(x,y)
plt.show()
```

```python
np.random.seed(0)
x1 = np.random.randint(0,100,200)
x2 = np.random.randint(0,500,200)
y=   np.random.randint(0,11,200)

fig,ax =plt.subplots()
ax.scatter(x1,x2,c=y)
plt.show()
```

## Distribution Plot
The distribution plot is suitable for comparing range and distribution for groups of numerical data. Data is plotted as value points along an axis.

### Histogram
A histogram is a plot that lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data. This allows the inspection of the data for its underlying distribution (e.g., normal distribution), outliers, skewness, etc.

Normally, `bins = sqrt(Number of Data Points)`

#### Histogram from a Continous Variable

```python

import math
np.random.seed(42)
X = np.random.randint(0,100,200)

bins = math.ceil(math.sqrt(X.size)) #or you can do bins='auto'

fig, ax = plt.subplots()

ax.hist(X,bins,color='#264827',alpha=0.5, rwidth=0.85)

ax.set_title('Histogram of Age')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')
ax.grid()

plt.show()
```

## Categorical Plots
Plots are basically used for visualizing the relationship between variables. Those variables can be either be completely numerical or a category like a group, class or division. In categorical plots, we try to plot our data according to the category as given.

### Bar Plot
A bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value

```python
X = ['Engineering Mathematics','Physics','C','Basic Electrical Engineering','Engineering Drawing','Applied Mechanics'] 
np.random.seed(42)
y= np.random.randint(32,81,6)

fig,ax = plt.subplots()
ax.bar(X,y,color='#264827',width=0.67)
ax.set_title('First Semester Marks')
ax.set_xlabel('Subjects')
ax.set_ylabel('Marks')
ax.grid()
plt.show()
```

```python
np.random.seed(42)

y = np.random.randint(30,51,8).reshape(2,4) 
X= np.arange(2073,2077)
print(y)
fig, ax = plt.subplots()

ax.bar(X, y[0] ,width=0.5, color='maroon')
ax.bar(X, y[1] , width=0.5,color='crimson',bottom=y[0])

ax.set_title('Money of both Male and Female')
ax.set_ylabel('Money (Rs.)')
ax.set_xlabel('Year')
ax.set_xticks(np.arange(2073,2077))
ax.legend(labels=['Male', 'Female'])
plt.show()
```

### Pie Chart
A Pie Chart can only display one series of data. Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie.

```python
X = ['Engineering Mathematics','Physics','C','Basic Electrical Engineering','Engineering Drawing','Applied Mechanics'] 
np.random.seed(42)
y= np.random.randint(32,81,6)

fig,ax = plt.subplots()
ax.pie(y,
       labels = X ,
       startangle=90,
        shadow= True,
        explode=(0,0.1,0,0.1,0,0.1),
        autopct='%1.1f%%')
ax.set_title('Marks Obtained in First Semester')
plt.show()
```
