# Seaborn

## What is Seaborn?
* Python data visualization library based on matplotlib
* library for making statistical graphics in Python
*  create more attractive and informative statistical graphics

## Why use Seaborn?
* builds on top of Matplotlib i.e if you know matplotlib you know half of seaborn already
* makes your traditional Matplotlib plots look a bit prettier.
* Library is relatively low-level in matplotlib while high-level in seaborn i.e follows declarative approach
* easier syntax than matplotlib
* Built in themes for styling matplotlib graphics
* generate more sophisticated visualizations such as heatmaps and pairplots  i.e close integration with pandas dataframe

**Importing our default libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

**Load some datasets**

```python
iris = sns.load_dataset('iris')
iris.head()
```

Lets visualize sepal_length vs sepal_width in both matplotlib first.

```python
#Creating a function to change categorical named data to numerical variable
def change_to_numerical(x):
  if(x == 'setosa'):
    return  1;
  elif (x == 'versicolor'):
    return 2;
  else:
    return 3;

iris_copy = iris.copy() 

iris_copy['species_num'] = iris_copy['species'].apply(change_to_numerical)

iris_copy.drop(['species'],axis=1, inplace=True)

iris_copy.head()


```

```python
X= iris_copy['sepal_length'].values
y= iris_copy['sepal_width'].values
species = iris_copy['species_num'].values

fig,ax=plt.subplots()

ax.scatter(X,y,c=species)

plt.show()
```

Now lets use seaborn

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
sns.set_style()
```

As you can see , we had to convert the non-numerical category to numerical before plotting in matplotlib and we did a lot of lines of code but in seaborn same thing could be achieved with 2 lines of code and a beautiful plot with many default things like label,grid and legend was built.

## Pairplot
Now we just checked the distribution of points with respect to one variable only but what if we need for every numerical independent variables.

In matplotlib, we would need to create various subplots and style them separately. It is so time consuming.

Lets see how we can achieve the same thing in seaborn.

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

Easy Peasy in just 2 lines of code

## Histogram

```python
# In Matplotlib
import math
np.random.seed(42)
X = iris['petal_length'].values

bins = math.ceil(math.sqrt(X.size)) #or you can do bins='auto'

fig, ax = plt.subplots()
ax.hist(X,bins,color='#264827',alpha=0.5, rwidth=0.8)
ax.set_title('Frequency of Petal Length')
ax.set_xlabel('Petal Length')
ax.set_ylabel('Frequency')
ax.grid(axis='both')
plt.show()
```

```python
#In Seaborn (Univariate)
sns.displot(iris,x='petal_length',binwidth=0.25,hue='species')
plt.show()
```

```python
#In Seaborn (Combining our Category)
sns.displot(iris,x='petal_length',kind='kde',hue='species')
plt.show()
```

## HeatMap
 Heatmaps represent the data in the form of a 2-dimensional format. Heatmaps visualize the data and represent in the form of a summary through the graph/colored maps.

 It uses various color palettes and different parameters to add more visualization features to the graph and thus adds to exploration of data effectively.

 Using heatmap to visaulize correlation is a better tool.

```python
sns.heatmap(iris.corr(),annot=True,cmap='summer')
plt.show()
```
